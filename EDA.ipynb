```python
import pandas as pd
from google.colab import files
import seaborn as sns 
import matplotlib.pyplot as plt
import io
import nltk
from wordcloud import WordCloud
from matplotlib import pyplot as plt
from nltk.corpus import stopwords
import collections 
```

```python
uploaded = files.upload()
```

```python
df_hate_speech = pd.read_csv(io.BytesIO(uploaded['hatespeech.csv']),lineterminator='\n')
```

```python
df_hate_speech.info()
```

```python
df_hate_speech['HateSpeech'].value_counts()
```

## Plots

```python
plt.figure()
sns.countplot(x='HateSpeech', data=df_hate_speech,hue='HateSpeech', alpha=0.9)
print(df_hate_speech['HateSpeech'].value_counts())
print(df_hate_speech['HateSpeech'].sum())
```

```python
data = df_hate_speech.groupby("HateSpeech").sum()
explode = (0.1, 0.0)
labels =['other','Hate Speech']
data.plot.pie(autopct="%.1f%%", title='Hate Speech', explode = explode,startangle=90,labels=labels,subplots=True)
```
